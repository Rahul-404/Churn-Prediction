# **Churn Prediction**

# Dataset Description

 Variable	| Definition 
 --- | --- 
customer_id | Unique ID for customer
vintage | vintage customer
age | age of customer
gender | Gender of customer
dependents | How many dependents customer have
occupation | Occupation of customer
city | city in which customer live
customer_nw_category | is customer from new category
branch_code | branch code of customer
current_balance | current balance of customer
previous_month_end_balance | previous month end balance  
average_monthly_balance_prevQ | average monthly balance for previous 1st quarter
average_monthly_balance_prevQ2 | average monthly balance for previous 2nd quarter
current_month_credit | current months credit money
previous_month_credit | previous months credit money 
current_month_debit |  current months debit money
previous_month_debit | previous months debit money
current_month_balance | current moths balance of customer
previous_month_balance | previous months balance of customer
churn | customer churn
last_transaction | last transaction date

# Evaluation Metric
The evaluation metric for this competition is F1 Score.
